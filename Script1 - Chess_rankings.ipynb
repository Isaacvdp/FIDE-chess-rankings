```python
# Import libraries

import pandas as pd
import numpy as np
import os
pwd = os.getcwd()
```

```python
# Load data

chess_rankings = pd.read_csv(pwd + "\Data\\Chess_FIDE_Rankings.csv")
grandmaster_list = pd.read_csv(pwd + "\Data\\chess_grandmaster_list.csv", encoding="latin1")
```

### Data cleaning and preliminary EDA

```python
chess_rankings.describe()
```

```python
chess_rankings.describe(include="object")
```

```python
chess_rankings["title"].unique()
```

```python
# Drop redundant columns

chess_rankings.drop(columns=["title"], inplace=True)
chess_rankings
```

```python
chess_rankings["federation "].mode()
```

```python
# Rename columns for consistency

chess_rankings = chess_rankings.rename(columns={"ELO ":"ELO", "federation ":"Federation", "rank ":"Rank", "name ":"Name", "games":"Games", "birth_year":"Birth Year"})
chess_rankings
```

```python
# Format datetime

from datetime import date
chess_rankings["Age"] = date.today().year - chess_rankings["Birth Year"]
```

```python
chess_rankings
```

```python
chess_rankings["ELO Difference"] = chess_rankings["ELO"].diff()
chess_rankings
```

```python
chess_rankings["ELO Difference"].describe()
```

```python
# Export finished data

chess_rankings.to_excel(pwd + "\Data\\FIDE_Chess_rankings_final.xlsx", index=False)
```
